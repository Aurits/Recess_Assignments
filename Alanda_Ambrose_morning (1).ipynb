```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Control Display of seaborn data
pd.options.display.max_columns=50
sns.set(style="darkgrid", rc={"axes.facecolor":"#CAF1DE"})
```

```python
df = pd.read_csv("bse_dataset.csv")
df
```

```python
# Understanding of the data
# 1.shape of the data
df.shape
```

```python
# 2.Checking column names
df.columns
```

```python
# 3.Checking for the information
df.info()
```

```python
# 4.Checking for null values
df.isnull().sum().sort_values(ascending=False)
```

```python
# 5.Checking null values in percentages
for i in df.columns:
    if df[i].isnull().any():
        print(i,"--------", df[i].isnull().sum()*100/df.shape[0],"%")
```

```python
# 7.Checking for duplicates
df.duplicated().sum()


# 8.Describe() statistical details about your data
df.describe().T
```
